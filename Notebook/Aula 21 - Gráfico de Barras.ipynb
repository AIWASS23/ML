```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/alissonSCA/dataset/master/Salary.csv')
```

```python
data.head()
```

# Descrição dos atributos

* Salário em milhares de dólares
* Gênero: 0-Feminino, 1-Masculino
* Idade em anos
* Phd: 0-Não, 1-Sim

```python
data['Gender'] = ['F' if x == 0 else 'M' for x in data['Gender']]
data['PhD'] = ['N' if x == 0 else 'S' for x in data['PhD']]

print(data.head())
```

```python
dt2 = data.groupby('Gender')['Salary'].mean()

print(dt2)
```

```python
plt.bar(dt2.index, dt2.values)
```

## Barra horizontal

* Nomes Grandes ou muitas barras

```python
plt.barh(dt2.index, dt2.values)
```

# Doutorado influencia?

A fonte da diferença pode ser o PhD?

```python
plt.bar(dt2.index, dt2.values, bottom = [0, 10])
```

```python
PHDs = np.unique(data['PhD'])

b = np.zeros(PHDs.size)
for phd in PHDs:
  dt2 = data.loc[data['PhD'] == phd, :].groupby('Gender')['Salary'].count()
  plt.bar(dt2.index, dt2.values, bottom = b)
  b = dt2.values

plt.legend(PHDs, title='PHD')
```

```python
dt2 = data.loc[data['PhD'] == 'S', :].groupby('Gender')['Salary'].mean()

plt.bar(dt2.index, dt2.values)
```

## Salários

```python
PHDs = np.unique(data['PhD'])

b = np.zeros(PHDs.size)
for phd in PHDs:
  dt2 = data.loc[data['PhD'] == phd, :].groupby('Gender')['Salary'].mean()
  plt.bar(dt2.index, dt2.values, bottom = b)
  b = dt2.values

plt.legend(PHDs, title='PHD')
```
