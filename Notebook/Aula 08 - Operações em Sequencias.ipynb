# Ordenar uma sequencia

* O método **sort** ordena os elementos de uma lista

```python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

# numbers.sort()

print(numbers)
```

# Cópia ordenada

* A função **sorted** retorna **uma cópia** da lista original, porém com os elementos ordenados.
  * A lista original **não é alterada**

```python
ordenada = sorted(numbers)

print(ordenada)
print(numbers)
```

# Buscar elemento

* Como encontrar um elemento em uma lista?

```python
print( 8 in numbers )

numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6, 8, 8, 0]

i = numbers.index(8)

print(i)
print(numbers[i])

print(numbers.count(8))
```

# Invertendo uma lista

```python
print(numbers)
numbers.reverse()
print(numbers)
```

```python
meus_numeros = numbers.copy()
meus_numeros[0] = -14
print(meus_numeros)
print(numbers)
```

# Listas bidimensionais (Matrizes)

```python
a = [[77, 68, 86, 73], [96, 87, 89, 81], [70, 90, 86, 81]]
```

AAHBDOV0_2.png

AAHBDOV0.png

```python
for i in range(3):
  for j in range(4):
    print(a[i][j], end='\t')
  print()
```

```python
for i, linha in enumerate(a):
  for j, coluna in enumerate(linha):
    print('a[%d][%d] = %d' % (i, j, coluna), end='\t')
  print()
```
