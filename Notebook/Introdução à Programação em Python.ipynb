# Aritimética

* soma (+)
* subtração (-)
* multiplicação (*)
* divisão (/)
* $a^b$ (**)
* Divisão Inteira
  * quociente (//)

```python
a = 2 + 2
b = 2 - 1
c = 3 * 4
d = 1 / 2
e = 2 ** 10

f = 16 // 3
g = 16 % 3
```

# Variáveis booleanas e o comando *if*

* $>$ $→$ $>$
* $<$ $→$ $<$
* $\geq$ $→$ $>=$
* $\leq$ $→$ $<=$
* $=$ $→$ $==$
* $\neq$ $→$ $!=$

```python
a = 7 == 8
print(a)
```

```python
if media >= 7:
  print('Aprovado!')

if media < 7:
  print('Reprovado')
```
