# Espaço Linear

Considere uma medicação pediátrica cuja posologia (em ml) em relação ao peso (em kg) é dada por 
$$\mathrm{pos}(p) = 4*\left(1.5 + \mathrm{tanh}(\frac{p - 15}{2})\right)$$

Imagine que você trabalha em um laboratório de farmacologia que pediu um gráfico para ser impresso na bula, com a posologia adequada a cada peso no intervalo entre 5 e 25kg, semelhante a imagem abaixo.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.linspace(5, 25, 100)
pos = 4 * (1.5 + np.tanh((p-15)/2) )
```

```python
plt.plot(p, pos, 'k-')
```

# Indexação - acessando elementos e intervalor

```python
from sklearn.datasets import load_iris
x, y = load_iris(return_X_y=True)
```

```python
print(y)
```

```python
a = x[:, [1, 0]]
print(a)
```

```python
b = x[:,0] > x.mean(axis=0)[0]
print(b.sum())

a = x[b, :]

print(a.shape)
```

```python
setosas = x[y==0, :]
versicolor = x[y==1, :]
virginicas = x[y==2, :]

print(setosas.shape)
print(versicolor.shape)
print(virginicas.shape)
```

# Exercício

Mostrar a média dos atributos das iris de cada tipo

```python
for i, tipo in enumerate(['setosa', 'versicolor', 'virginica']):
  print(tipo)
  flores = x[y==i, :]
  print(flores.mean(axis=0))
```
