# Gráfico de linha

* dados ordenados vs dados quantitativos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/alissonSCA/dataset/master/PoliceDeathsInAmerica.csv')
```

```python
data.head()
```

```python
dt2 = data.groupby('year')['index'].count()
```

```python
plt.plot(dt2.index, dt2.values, 'k-')

plt.xlabel('ano')
plt.ylabel('# mortes')
```

```python
np.unique(data['cause_short'])
```

```python
gFire = data['cause_short'] == 'Gunfire'
```

```python
dt2 = data.loc[gFire, :].groupby('year')['index'].count()
plt.plot(dt2.index, dt2.values, 'k-', alpha=0.8)

dt2 = data.loc[~gFire, :].groupby('year')['index'].count()
plt.plot(dt2.index, dt2.values, 'r-', alpha=0.8)


plt.xlabel('ano')
plt.ylabel('# mortes')

plt.legend(['Sim', 'Não'], title='Arma de fogo')
```
