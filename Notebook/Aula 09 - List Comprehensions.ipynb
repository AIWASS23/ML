# Criação Simplificada de Listas

```python
list(range(6))
```

```python
lista = []
for valor in range(6):
  lista.append(2*valor)
print(lista)
```

```python
lista2 = [ valor**2 for valor in range(6) if valor % 2 == 0]
print(lista2)
```

```python
def primo(n):
  b = True
  for divisor in range(2, n):
    if n % divisor == 0:
      b = False
  return b

```

```python
primos = [x for x in range(1, 101) if primo(x)] # lista dos números primos entre 201 e 300
print(len(primos))
```

## Filtro e Mapeamento de Valores

```python
numeros = list(range(1, 101))

primos = list(filter(primo, numeros))
print(len(primos))
```

```python
notas = [10, 8, 6, 6.2, 9, 2, 4, 8]

def esta_na_final(nota):
  return nota < 7 and nota >= 4

af = list(filter(esta_na_final, notas))

print(len(af))
```

```python
novas_notas = list(map(lambda nota: nota+1, notas))

print(novas_notas)
```

## Funções Lambda

```python
af = list(filter(lambda nota: nota < 7 and nota >= 4, notas))

print(len(af))
```
