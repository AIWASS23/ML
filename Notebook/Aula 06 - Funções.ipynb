# Parâmetros das funções

* multiplos parâmetros
* valores padrão
* identificação na chamada
* entrada como sequencia

```python
def eleva(base = 2.0, expoente = 1.0):
  return base**expoente

print( eleva(base = 2, expoente = 10) )
```

```python
def somar(*x):
  total = 0
  for parametro in x:
    total += parametro
  return total

print(somar(1, 2, -12))
```

# Números aleatórios

```python
import random
random.seed(-1)


quantidades = [0, 0, 0, 0, 0, 0]
nVezes = 60000

for i in range(nVezes):
  valor_sorteado = random.randrange(1, 7)
  quantidades[valor_sorteado - 1] += 1

print(quantidades)


```

# Reprodutibilidade (semente)

```python
random.seed(32)

for i in range(10):
  valor_sorteado = random.randrange(1, 7)
  print(valor_sorteado)

# 1, 2, 2, 3, 6, 2, 4, 1, 6, 1
```
