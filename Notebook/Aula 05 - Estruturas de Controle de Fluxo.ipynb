# Estruturas de Controle de Fluxo

```python
nota = 5.9
if nota >= 7:
  print('Aprovado')
print('Concluido')  
```

```python
nota = 3.8
if nota >= 7:
  print('Aprovado')
else:
  print('Reprovado')
print('Concluído')
```

```python
nota = 2

if (nota >= 7):
  print('Aprovada')
elif (nota >= 3):
  print('AF')
else:
  print('Reprovado')

print('Concluido')
```

# Estruturas de Repetição

**Exemplo:** Somar os números naturais entre 1 e $n$

```python
n = 4

i = 1
soma = 0
while (i <= n):
  soma += i
  i += 1

print(soma)
```

```python
a = 2
b = 10


a **= b
print(a)
```

# Listas

```python
#     0     1      2
a = [-1, 3.14, 'Aula']

for valor in a:
  print(valor)
```

```python
n = 3

soma = n
for valor in range(n):
  soma += valor
print(soma)
```

# Exemplo: Calcular a média da turma

**Caso 1:** valores fixos em uma lista

```python
notas = [10, 8.9, 7.3, 5.4, 8.3, 6.9, 9.8, 8.8, 7, 9.9]

qtd = 0
soma = 0
for nota in notas:
  soma += nota
  qtd += 1

print('Média = ', soma/qtd)
```

**Caso 2:** Sentinela

```python
soma = 0
qtd = 0

while True:
  nota = float(input('Digite a nota (-1 para concluir): '))
  if (nota == -1):
    break
  if (nota < 0):
    print('Nota inválida!')
    continue
  soma += nota
  qtd += 1
  
print('Média = ', soma/qtd)
```
